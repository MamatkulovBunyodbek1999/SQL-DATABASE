# CHANGING DATAS WITH .map()

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

 # ['num-of-cylinders'] COLUMN HAS STRING DATAS, LET'S CHAN IT INTO INTEGER 

```python
df['num-of-cylinders'].unique()
```

```python
dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12}
```

# Now we will add a new column [cylinders] into dataframe 

```python
df['cylinders'] = df['num-of-cylinders'].map(dict)
```

```python
df.head()
```

# LET'S EXECUTE THOSE TWO COLUMNS TO SEE THE DIFFERENCES

```python
df[['num-of-cylinders','cylinders']]
```

# LET'S CONVERT USD PRICES INTO UZS WITH MAPPING METHOD

```python
kurs=10720 # 1$=10720 so'm
df['price'].map(lambda x: x*kurs)
```

# IF WE WANT, WE CAN ADD IT INTO OUR DATAFRAME

```python
df['price_uzs'] = df['price'].map(lambda x: x*kurs)
df.head()
```

# REPLACING DATA

```python
df.replace(['four','six'],[4,6])
```

# Replacing index and  column values

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population_duplicates.csv', index_col='country')
```

```python
df2.head()
```

# .Rename()
converting indexes and columns into uppercase


```python
df2.rename(index=str.upper, columns=str.upper) # If we need we can use inplace=True method
```

# .replace can help us to replace inex names and column names as well

```python
df2.rename(index={'Macau':'MAKAO'}, columns={'pop2021':'population'})
```
