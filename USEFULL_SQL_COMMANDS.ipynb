# COMMANDS IN SQL 
# DISTINCT

```python
query = "SELECT DISTINCT country FROM elect"
countries = pd.read_sql(query, db)
countries.head(10)
```

# CHECKING IF THERE IS A CERTAIN ELEMENT IN A CERTAIN COLUMN 
IN OUR CASE ELEMENT IS UZBEKISTAN, COLUMN= COUNTRY

```python
'Uzbekistan' in countries['country'].values
```

# LET'S SEPARATE DATAS WHICH ARE BELONG TO UZBEKISTAN

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan'"
dfuz = pd.read_sql(query,db)
dfuz.head()
```

# LET'S SEPARATE DATAS WHICH ARE BELONG TO UZBEKISTAN BEFORE 2010

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year>2010"
dfuz = pd.read_sql(query,db)
dfuz
```

# FILTERING WITH [BETWEEN]
"SELECT * FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2010 AND 2021

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2010 AND 2021"
pd.read_sql(query, db)
```

FILTERING WITH [ IN ]
"SELECT * FROM elect WHERE country='Uzbekistan' AND year IN (1990,1995,2000,2005,2010)"

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year IN (1990,1995,2000,2005,2010)"
pd.read_sql(query, db)
```

# FILTERING WITH [ LIKE ]
%stan ---> it will execute all data ending with 'stan'



_tol ----> it will execute all data ending with 'tol'. everything depends on the location and amount under line sign (  _ ) is

```python
query = "SELECT * FROM elect WHERE country LIKE '%stan' AND year IN (2019, 2020, 2021)"
pd.read_sql(query, db)
```

## OR

```python
query = "SELECT * FROM elect WHERE code LIKE 'A__' AND year=2020"
pd.read_sql(query, db)
```
