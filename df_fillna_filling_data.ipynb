# Its better to re-fill data if possible rather than dropping

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv", index_col=0)
df.head()
```

converting all nan values into 0

```python
df.fillna(0)
```

# We can convert NaN values into standart values, according to a certain column's common values

```python
df.fillna({'wheel-base':90, 'horsepower':100})
```

# .ffil>>>  forwardfill.
# when we have similar values in a column, we can use this method. 
df[df.columnName=='something'].fillna(method='ffil')

```python
df[df.company=='audi'].fillna(method='ffill')
```

##  method='bfill' BACKFILL

```python
df[df.company=='toyota']
```

```python
df[df.company=='toyota'].fillna(method='bfill')
```

```python
dfh = df[df['body-style']=='hatchback']
dfh
```

```python
dfh.fillna(dfh.mean())
```

# .(fillna) according to a certain columns mean values !

```python
df.fillna({'horsepower':np.mean(df['horsepower']), "wheel-base":np.mean(df['wheel-base'])})
```
